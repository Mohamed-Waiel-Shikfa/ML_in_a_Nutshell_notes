***

If we use <span class="girk">linear regression</span> to predict the price of a house, the prediction will be  based on the <u><span class="mark">linear combination of the features</span></u> (i.e., every feature contributes additively to the price):
<br><br>
    
<span style='color:blue; font-size:110%'>
    
$$ \color{green}{price}\ = \sum_{i\ \in\ |All features|} w(i)\cdot \color{red}{feature(i)} $$ 

</span>

where the coefficients $w(i)$ have to be learned. 

***
The fact that for instance feature CRIM ranges from 0.006 to 0.06, while TAX features ranges from 222 to 294 potentially poses a problem since a 1% change in TAX has potentially a much larger impact in the sum that a 1% change in CRIM. 

<font color='red'>$\to$ A ML algorithm could be deceived, naturally assigning more importance to TAX than to CRIM, which might not be necessarily the case.</font>.

Note that, in particular, OLSR is NOT affected by the difference in scale. Other algorthms for linear regression will be affected, instead.

***
<span style='color:blue; font-size:110%'>
    
$\to$ We might want to better **balance and equalize feature ranges** in order not to give an unwanted feature more weight than to some of the other features $\rightarrow$ **<font color="orange">Rescale the features</span>**
    
</span>

# <font color='red'>Scaling features using `sklearn`

To (re)scale features to convenient and balanced ranges, `sklearn` offers a number of methods.

***
## <font color='red'>Standardization: $z$-transform

<font color='blue'>**Standardization** </font> consists in transforming the feature vectors such that different features vectors have <font color="orange">**comparable** mean and spread</font> (i.e. standard deviation). 

The <span class="birk">z-transform</span> obtains this by converting the values of a sample into <span class="pirk">**z-scores.**</span> This results into a sample with <span class="mark">zero mean</span> and <span class="mark">unit variance.</span> 

***
In practice we can ignore the shape of the distribution (maybe we have no idea about it!) and just <span class="mark">transform the data</span> to **<span class="girk">center** them</span> by removing the mean value of each feature, then <span class="girk">**scale** the data</span> by dividing non-constant features by their standard deviation.

In other words: 
<span style='color:blue; font-size:110%'>
**standardization means to <span class="mark">apply a $z(x)$ transform to each feature value $x$!</span>**
<br><br>
$$z(x) = \frac{x-\mu}{\sigma}$$
</span>

$\to$ **The resulting data have <span class="girk">zero mean</span> and <span class="pirk">unit standard deviation</span>**

Note: this is an operation that applies to each individual feature! (where each feature has its own mean and standard deviation)

***
- **Linear transformation:** Z-transform is a <u>linear transformation</u> that <span class="mark">doesn't change the shape of the data</span>

***
- If $\mu$ or $\sigma$ are not known (e.g., by prior knowledge or based on model assumptions), they can be substituted by their **<span class="girk">sample estimates</span>**, computed directly on the sample data.

***
## <font color='red'>Example: Apply a z-transform to a <u>random uniform</u> data sample

```python
import scipy.stats as stats
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['figure.dpi']= 100 # set the resolution to x dpi
```

First, generate and plot uniformly distributed data 

```python
x_low = -3
x_high = 5
sample_size = 2000

data_uniform = np.random.uniform(x_low, x_high, sample_size)
```

Mean and standard deviation of the sample data

```python
data_uniform.mean(), data_uniform.std()
```

```python
fig = plt.figure(figsize=(10, 5))

nbins = 40

plt.hist(data_uniform, bins=nbins, density=True, color='blue', 
         label='Uniform', alpha=0.99)
plt.legend()
plt.show()
```

***
<span class="girk">Apply a z-transform</span> and show the result

```python
z_transformed = (data_uniform - data_uniform.mean()) / data_uniform.std()
```

New mean and standard deviation

```python
z_transformed.mean(), z_transformed.std()
```

```python
#fig = plt.figure(figsize=(10, 5))

plt.hist(z_transformed, nbins, density=True, color='orange', 
         label='Z-transformed uniform data', alpha=0.8)
plt.legend()
plt.show()
```

***
Let's plot them side by side (to also learn how to make a multiple plot in a matrix!)

```python
# create a figure, that will host two suplots
# organized in 1 row, with 2 plots

fig, sbplots = plt.subplots(1, 2, figsize=(10, 3))

sbplots[0].hist(data_uniform, nbins, density=True, color='blue', 
            label='Uniform original', alpha=0.8)
sbplots[0].set_xlabel('$x$')
sbplots[0].set_ylabel('Probability density')
sbplots[0].legend(loc='lower left')

sbplots[1].hist(z_transformed, nbins, density=True, color='orange', 
            label='z-transformed', alpha=0.9)
sbplots[1].set_xlabel('$x$')
sbplots[1].set_ylabel('Probability density')
sbplots[1].legend(loc='lower right')

fig.subplots_adjust(wspace=0.4)


plt.show()
```

## <font color='red'> `scale()` method for implementing a $z$-transform
    
<font color='blue'>The `scale()` method from the `preprocessing` module of  `sklearn`
    performs the <span class="mark">$z$-transform</span> on the array data.</font>
    
It uses <span class="mark">sample estimates</span> of mean and standard deviation.

```python
from sklearn import preprocessing
import numpy as np

X_train = np.array([ [ 1., -1.,  2.],
                     [ 2.,  0.,  4.],
                     [ 0.,  1., -1.] ])

X_train.mean(), X_train.std()
```

```python
X_scaled = preprocessing.scale(X_train)

X_scaled
```

```python
#help(preprocessing.scale)
```

***
By default, `axis=0` $\to$ <span class="mark">Scaled data has zero mean and unit variance</span> <font color='red'>**along each feature column**</font> and, accordingly, for the whole data matrix (taken as one flat array)

```python
X_scaled[:, 0].mean(), X_scaled[:, 0].std()
```

```python
X_scaled[:, 1].mean(), X_scaled[:, 1].std()
```

```python
X_scaled[:, 2].mean(), X_scaled[:, 2].std()
```

```python
print(X_scaled.mean())

print(X_scaled.std())
```

***
Note that <span class="mark">each feature / column ranges between some minimum and maximum values that differ from feature to feature.</span> 

$\to$ The ranges of the features differ but <span class="mark">their statistical properties are the same</span>, since <font color='orange'>all features are now represented as *standardized* random variables.</font> 

***
**Rows are not standardized!**

```python
X_scaled[0, :].mean(), X_scaled[0, :].std()
```

... but we can change the axis to make the standardization along the rows

```python
X_scaled = preprocessing.scale(X_train, axis=1)

X_scaled
```

```python
X_scaled[0, :].mean(), X_scaled[0, :].std()
```

## <font color='red'>`StandardScaler` to (re)apply the transformation to a testing set 

The preprocessing module further provides a <font color='blue'> utility class `StandardScaler` that computes mean and standard deviation on a **training set** and <span class="mark">stores it</span> so as to be able to later reapply the same transformation on a **testing set.** </font>

```python
X_train = np.array([[ 1., -1.,  2.],
                     [ 2.,  0.,  4.],
                     [ 0.,  1., -1.]])

scaler = preprocessing.StandardScaler().fit(X_train)
type(scaler)
```

```python
#help(preprocessing._data.StandardScaler)
```

```python
X_train
```

`scaler` is an instance of a `StandardScaler` object that <span class="girk">computes and stores the mean and the standard deviation of the data</span>, but <span class="pirk">it does nothing (i.e., **it doesn't fit the data**</span>).

<font color='red'> The attributes `mean_` and `std_` contain the computed statistics.</font>

```python
print(scaler.mean_)

print(scaler.scale_)
```

***
<font color='red'> $\to$ The <u>stored transformation</u> is <span class="mark">**applied to the data**</span> (i.e., training set) only with the call of the `transform()` method.</font>

```python
X_scaled = scaler.transform(X_train)
```

```python
X_scaled
```

<font color='blue'>$\to$ The scaler instance <span class="girk">can then be used on **new data**</span> (e.g., test data) to transform it the same way it did on the training set. </font>.

```python
X_test = [[-1., 1., 0.]]
X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

***
<span class="pirk">It is possible to disable either centering or scaling</span> by either passing `with_mean=False` or `with_std=False` to the constructor of the `StandardScaler` object.

## <font color='red'> Apply z-transform to the Wisconsin Breast Cancer dataset: k-NN performance

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7642)
```

***
Create K-NN classifier **without feature scaling**

```python
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train, y_train)
y_pred_no_scaling = knn_classifier.predict(X_test)
```

```python
# Calculate accuracy without feature scaling

accuracy_no_scaling = accuracy_score(y_test, y_pred_no_scaling)
print("Accuracy without feature scaling: {:.3f}".format(accuracy_no_scaling))
```

***
Create K-NN classifier **with feature scaling**

```python
scaler = preprocessing.StandardScaler().fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn_classifier_scaled = KNeighborsClassifier()

knn_classifier_scaled.fit(X_train_scaled, y_train)
y_pred_scaling = knn_classifier_scaled.predict(X_test_scaled)

```

```python
# Calculate accuracy with feature scaling
accuracy_scaling = accuracy_score(y_test, y_pred_scaling)
print("Accuracy with feature scaling: {:.3f}".format(accuracy_scaling))
```

# <font color='red'>Scaling to a range: `MinMaxScaler`

An alternative to standardization is scaling features to lie between a<font color='blue'> **given minimum and maximum values.** </font>

$\to$ The motivations to use this type of scaling include robustness to very small standard deviations of features and preserving zero entries in sparse data. This is also a way to have <span class="mark">all the features ranging in the same interval of values such that they can be well balanced.</span>

The main disadvantage of this type of transformation is that the <u>statistical properties of the data aren't taken into account,</u> while this is the case for the other transformations. 

***
The *standard* way of scaling between a minimum and maximum value consists in scaling <font color='blue'>**between zero and one.**</font> This is also commonly referred to as <font color='red'><u><span class="girk">Normalization.</u> </font></span>

<span class="burk">Warning</span>: the term *normalization* is given different meanings depending on the different scientific domain and the different context of application. Below we'll see that in `sklearn` the term normalization is referred to the case of creating a **unit-norm feature vector.**

***
The <span class="girk">scaling between a minimum and a maximum</span> value can be performed using the `MinMaxScaler` object.

<span style='color:blue; font-size:110%'>
<br>  
    
**As default, it performs the scaling to <span class="girk">$[0,1]$</span>,** i.e., it scales the entry <span class="mark">$x_i$</span> of feature array <span class="pirk">$x$</span> as: <br><p>
    $$MMS(x_i) = \frac{\color{red}{x_i} - min(x)}{max(x) - min(x)}$$
   
</span>

Note that the $[0,1]$ scaling can be used to **rescale between any desired range** <span class="pirk">$[a,b]$</span> by: <br><p>
    <font color='red' size=+1.45>$$a + (b-a)\cdot MMS(x)$$</font>
***

Example scaling between $[0,1]$ using `MinMaxScaler`:

```python
X_train = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  4.],
                    [ 0.,  1., -1.]])

# get the scaler object
min_max_scaler = preprocessing.MinMaxScaler().fit(X_train)
```

```python
# apply the scaler to the data, columnwise
X_train_minmax = min_max_scaler.transform(X_train)

X_train_minmax
```

***
## <font color='red'>Fit and transform in one single operation: `.fit_transform()` 
    
`preprocessing` also offers the possibility to **<span class="mark">fit and transform in one single operation</span>** starting from an empty scaler object, as it is shown below using `.fit_transform()` (but using the previous approach is cleaner)

```python
X_train = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  4.],
                    [ 0.,  1., -1.]])

# get the scaler object
min_max_scaler = preprocessing.MinMaxScaler()

# apply the scaler to the data
X_train_minmax = min_max_scaler.fit_transform(X_train)

X_train_minmax
```

***
Again, the same instance of the transformer can then be applied to some **new test data** unseen during the fit call: <font color='red'>the same scaling and shifting operations will be applied to be consistent with the transformation performed on the train data.</font>

```python
X_test = np.array([[-3., -1.,  4.]])

# transform with the same scaler used for the training data
X_test_minmax = min_max_scaler.transform(X_test)

X_test_minmax
```

$\to$ Note that the **<span class="mark">test values are not in $[0,1]$.** This is consistent, since the applied scaling was based on the value ranges from the training set (that might not be entirely representative of the ranges in the test set).
</span>
***

An explicit `feature_range=(min, max)` can be used as argument of the `MinMaxScaler` constructor in order to **<span class="girk">choose any desired range.</span>**


```python
X_train = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  4.],
                    [ 0.,  1., -1.]])

# get the scaler object
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-0.5, 0.5))

# apply the scaler to the data
X_train_minmax = min_max_scaler.fit_transform(X_train)

X_train_minmax
```

# <font color='red'>Scale to a max absolute value: `MaxAbsScaler`

Scale so that the **maximum absolute value of each feature is scaled to unit size.** 

The `MaxAbsScaler` scaler object does the job. 

<span style='color:blue; font-size:110%'>
    
`MaxAbsScaler` works in a very similar fashion as the `MinMaxScaler`, but scales in a way that the **training data lies within the range [-1, 1]** by dividing through the largest maximum value in each feature. 

</span>
    
<span class="mark">It is meant for data that is already centered at zero or for sparse data.</span>

```python
X_train = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  4.],
                    [ 0.,  1., -1.]])

max_abs_scaler = preprocessing.MaxAbsScaler()

X_train_maxabs = max_abs_scaler.fit_transform(X_train)
X_train_maxabs
```

```python
X_test = np.array([[ -3., -1.,  4.]])
X_test_maxabs = max_abs_scaler.transform(X_test)
X_test_maxabs
```

$\to$ Note that the **data in the test set aren't between [-1,1]** because the max absolute value from the training set has been used by the transform, and not necessarily the max absolute values are the same in the two datasets.

# <font color='red'>Normalize to scale to a unit-norm vector: `Normalizer`

As pointed out above, the term normalization can be interpreted in different ways. In `sklearn` the `Normalizer` object normalizes the samples individually to unit-norm, depending on the chosen norm (either L2, which is the Euclidean one, or L1, which is the abs norm).<br><br>

<span style='color:blue; font-size:110%'>
    
Given a list of feature values $x_i, i=0, \ldots n$, under the L2 norm, the effect of the normalization is to rescale each value such that $$\sum_{i=0}^n x_i^2 = 1,$$ which makes the <span class="girk">Euclidean norm of the normalized feature vector equal to one:</span> $||x|| = \sqrt{\sum_{i=0}^n x_i^2} = 1.$

$\to$ By this normalization, **feature values are <span class="mark">necessarily shrunk into some subset of $[-1,1]$</span> in order to meet the unit-norm constraint.**
</span>

***

```python
X = np.array([[4, 1, 2, 2],
              [1, 3, 9, 3],
              [5, 7, 5, 1]])

normalizer = preprocessing.Normalizer().fit(X)  # fit does nothing
normalizer
```

```python
Xn = normalizer.transform(X)
```

```python
Xn
```

$\to$ Note that in this case by default the normalization happens along the axis 1: each row / example gets normalized to a unit vector

```python
np.sum(Xn[0, :]**2)
```

```python
np.sum(Xn[:, 0]**2)
```

<span class="mark">To make normalization being performed on the columns (features!)</span>, data is given reversing columns in to rows and vice versa.

On matrices, this operation is called **Transposition.**

Of course, pandas and numpy make transposition easy :-)

```python
#XT = np.transpose(X)
XT = X.T # transposed matrix
XT
```

```python
normalizer = preprocessing.Normalizer().fit(X.T) 
normalizer
```

```python
 # transform row by row, but now on transposed
XnT = normalizer.transform(X.T) 
```

```python
# get back the original data, normalized, by re-transposing the matrix
Xn = XnT.T
print(XnT.T)
np.sum(Xn[:, 0]**2)
```

***
Another way of obtaining the same result is by using the `normalize()` method, as it is shown below, specifying `axis=0` in order to go column-wise performing the normalization.

```python
Xn = preprocessing.normalize(X, norm='l2', axis=0)
Xn
```

```python
np.sum(Xn[:, 0]**2)
```

# <font color='red'>Robust scaling: `RobustScaler()`

The disadvantage of min-max scaling is that often the <span class="mark">presence of **outliers** affects the scaled values for any feature.</span><br><br>

<span style='color:blue; font-size:110%'>
    
**Robust scaling** tries to account for this fact by using a statistical measure to scale features without being affected by outliers
    
$$RS(x_i) = \frac{x_i - \color{red}{median(x)}}{\color{green}{IQR(x)}}$$
</span>

The scaler object `RobustScaler()` performs the robust scaling:

```python
X_train = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  14.],
                    [ 0.,  1., -1.]])

# get the scaler object
rs_scaler = preprocessing.RobustScaler()

# apply the scaler to the data
X_train_robust = rs_scaler.fit_transform(X_train)

X_train_robust
```

```python
X_test = np.array([[ -3., -1.,  4.]])
X_test_rs = rs_scaler.transform(X_test)
X_test_rs
```
